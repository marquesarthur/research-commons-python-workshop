## Adding new variables

#### Adding column representing the year

```python
mydata['year'] = mydata.apply(lambda row: datetime.strptime(row['dt'], '%Y-%m-%d').year, axis = 1)
```

**Whait? What?**

Now that we now how to convert a single entry, let's do that for every element in our table

```python
# The lambda is like a loop. You can read it like:
# for every row in my data apply the datetime.strptime function ...
# ... which converts the element to a date and we extract the year of that date

# the [:100] is just to speed up the output.

mydata[:100].apply(lambda row: datetime.strptime(row['dt'], '%Y-%m-%d').year, axis = 1)
```

```python
# Finally, we create a new column to store the data containing the year
mydata['year'] = mydata.apply(lambda row: datetime.strptime(row['dt'], '%Y-%m-%d').year, axis = 1)
```

## Some housekeeping and background

### We are merging the tables by year!

```python
carbon_df = carbon.groupby(['year']).agg({'CarbonDioxide' : 'mean'})
carbon_df = carbon_df.reset_index()
temperature_df = mydata.groupby(['year']).agg({'AverageTemperature' : 'mean'})
temperature_df = temperature_df.reset_index()
new_data = pd.merge(temperature_df, carbon_df, on='year')
```

```python
new_data.head()
```

### Hypothesis testing

From [Wikipedia](https://en.wikipedia.org/wiki/Statistical_significance):

#### To determine whether a result is statistically significant, a researcher calculates a p-value, which is the probability of observing an effect of the same magnitude or more extreme given that the null hypothesis is true.

#### The null hypothesis is rejected if the p-value is less than a predetermined level, α.

#### α is called the significance level, and is the probability of rejecting the null hypothesis given that it is true (a type I error). 

#### α is usually set at or below 5%.

<img src="figures/normal_curve.png">

## Our null hypotheses

### There is no difference in the Average temperature in the `gas & oil` era and the `electronic` era

```python
mydata.head()
```

#### Independent T-test compares means between two groups

It is often used to see whether there is a group difference in continuous data **between two groups** 

*Model assumptions*

    (1) Independence

    (2) Normality

    (3) Equal variance

```python
group_a = mydata[mydata["era"] == "gas & oil"]["AverageTemperature"]
group_b = mydata[mydata["era"] == "electronic"]["AverageTemperature"]

stats.ttest_ind(group_a, group_b, equal_var = True)
```

Interpreting the results:

* `p-value < 0.0` pretty low, so that means that we can reject the null hypothesis

<br>

* Which one seems higher?

```python
print("gas & oil =", np.mean(group_a))
```

```python
print("electronic =", np.mean(group_b))
```

### Is there any association between the `AverageTemperature` and the `AverageCarbonEmission`  ?

#### Pearson’s correlation is used to examine associations between variables (represented by continuous data) by looking at the direction and strength of the associations

```python
new_data.head()
```

```python
corr, pvalue = stats.pearsonr(new_data['AverageTemperature'], new_data['CarbonDioxide'])
```

### Interpreting the results:

```python
print("p-value=", pvalue)
```

that means that there is statistically significant correlation between `temperature` and `carbon emission`

#### How strong is the correlation?

```python
print("corr = ", corr)
```

`cor` varies from `-1` to `1` positive values indicate that an increase in the `y` variable increases the `x` variable by `corr`.
